```python
import pandas as pd ;import numpy as np
```

```python
v1 = pd.DataFrame({"anahtar":["a","b","c","c","d","e"],"sayi1":range(6)})
v2= pd.DataFrame({"anahtar":["b","c","d","e"],"say2":range(4)})
```

```python
v1
```

```python
v2
```

```python
#merge methodu ile birlestiriyoruz
#anahtar kelimesini PK olarak aldi 
pd.merge(v1,v2)
```

```python
#istedigimiz satira gore birlestime de yaapbilriz
#anahtar sutunu iki veride de ayni oldguundan rakam_x ve rakam_
pd.merge(v1,v2, on="anahtar", how= "outer")
```

```python
v3 = pd.DataFrame({"anahtar1":["a","b","c","c","d","e"],"sayi1":range(6)})
v4= pd.DataFrame({"anahtar2":["b","c","d","e"],"say2":range(4)})
```

```python
#farkli isimlerdeki sutunlari da birlestirebiliriz
#how=outer kullnanarak eksik verileri NaN olarak sileme yazdirabiliirz 
#how in yanina left ya da right,inner vs de yazilabilir
pd.merge(v3,v4,left_on="anahtar1",right_on="anahtar2",how="outer")
```

```python
v1 = pd.DataFrame({"anahtar":["a","b","c","c","d","e"],"sayi1":range(6),
                   "rakam":["bir","uc","iki","bir","bir","iki"]})
v2= pd.DataFrame({"anahtar":["b","c","d","f"],"say2":range(4),
                  "rakam":["bir","bir","iki","iki"]})
```

```python
v1
```

```python
v2
```

```python
#anahtar ve rakam degerlirine gore 2 datayi birlestirdik
#how = outer ile NaN Ddegerleri de yazdirdik 
pd.merge(v1,v2,on=["anahtar", "rakam"],how = "outer")
```

```python
#2 erideki anahtar bilgilerini yazdirdik, sonra da  otomatrik atanan isimleri degistirdik
#merge , on , how ve suffixes kullandik
pd.merge(v1,v2, on="anahtar", how="outer",suffixes=["_veri1","_veri2"])
```

```python
#birlestiemede indexlere 
#left_index=True ya da rightIndex=True

df1 = pd.DataFrame({"harf":["a","a","b","b","a","c"],
                    "say":range(6)})
df2 = pd.DataFrame({"deger":[3,5,7]},index =["a","b","e"])
```

```python
df1
```

```python
df2
```

```python
#df2 deki "deger sutununu, df1 in sag tarafina ekledik. left_on=harf diyerek harfin en solda olacagini belirttik 
pd.merge(df1,df2,left_on="harf",right_index=True)
```

```python
sag = pd.DataFrame([[1,2],[3,4],[5,6]],index = ["a","c","d"], columns = ["Ali", "Nur"])
sol = pd.DataFrame([[7,8],[9,10],[11,12],[13,14]], index = ["a","b","e","f"], columns = ["Efe","Ata"])
```

```python
sag
```

```python
sol
```

```python
#hem left hem right index i True olarak kullandik
#how = Outer ile yine eksik verileri NaN olarak gosterdik
pd.merge(sag,sol,right_index=True,left_index=True, how="outer")
```

```python
#join methodu 
sol.join(sag, how="outer")
```

```python
#concat methodu 
#dizilerin birlestirilmesi ( ayni boyuttaki iki diziyi birlestirmek
dizi = np.arange(20).reshape(4,5)
dizi
```

```python
#kendisiyle birlestirelim (sutun olarak ) 
np.concatenate([dizi,dizi],axis= 1)
```

```python
#satir olarak 
np.concatenate([dizi,dizi],axis= 0)
```

```python
#farkli indexlere sahip verileri birlestirmek icin concat kullanilabilir
veri1 = pd.Series([0,1], index = ["a","b"])
veri2 = pd.Series([2,3,4], index = ["c","d","e"])
veri3 = pd.Series([5,6], index = ["f","g"])
```

```python
veri1
```

```python
veri2
```

```python
veri3
```

```python
#3 veriyi birlestirdik, sutuna gore , join inner ile kesisim kumelerine baktik ama ortak ozelliklero yok 
pd.concat([veri1,veri2,veri3],axis = 1 , sort = False , join = "inner")
```

```python
#yeni bir veri olusturduk ve kesisim kumelerinde ortak veri var 
#join_axes ile istedigimiz veriyi listeye manuel olarak ekleyebbiliriz
veri4 = pd.Series([10,11,12], index =["a","b","c"])
pd.concat([veri1,veri4], axis = 1 , join ="inner")
```

```python
#keys secenegi ile verilere index de atadik , sutuna gore 
x = pd.concat([veri1,veri2,veri4],axis = 1 , keys = ["bir","iki","uc"], sort= False)
x
```

```python
#indexi olmayan verileri birlestirme
#ignore_index = True. bunu yazmazsak hata verir 

veri1 = pd.DataFrame(np.random.randn(3,4), columns = ["a","b","c","d"])
veri2 = pd.DataFrame(np.random.randn(2,3), columns = ["b","d","a"])
pd.concat([veri1,veri2],ignore_index = True, sort= False)
```
