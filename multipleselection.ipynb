```python
#sutun nasil pratik olarak secilir 
#loc methodu
#iloc methodu 
#coklu secme
#coklu filtreleme
import pandas as pd 
```

```python
film = pd.read_csv("http://bit.ly/imdbratings")
film
```

```python
film.head()
```

```python
#selection , farki sekillerde yapilabilir / koslei parantez 
#1den fazla sutun gostermek istersek , 2 tane koseli parantez icine yazilir 
film[["title","genre","duration"]].head()
```

```python
#loc methodu (ilk deger sutun, 2. deger satiei gosterecek ) 
#0 yazip bosluk birakirsak 1 numarali filmi gosterecek 
#1 den fazla secmek icin yine 2 tane koseli parantez 
film.loc[[0,5,10],]
```

```python
#ilk 5 filmin title ve genresi
film.loc[0:5,["title","genre"]]
```

```python
#title den genreye kadar tun sutunlari aliyor , koseli parantez kalkti 
film.loc[0:5,"star_rating":"duration"]
```

```python
#ilk satiri bos birakacak da bu islem yapilabilir 
#loc her zaman sutunlari gosteriyor 
film.loc[:,"title":"genre":]
```

```python
#loc ile suzme islemi
#sadece Crime turundeki filmleri goster
#sona viegul koymak onemli
film.loc[film.genre=="Crime",]
```

```python
#sadece title sutunu yazdirmak, 
film.loc[film.genre=="Crime",["title","duration"]]
```

```python
#iloc methodu , loc tan farki, satir ve sutunlarin indexlerini gosterir
#bu veride 0. index degeri "star_rating" o yuzden sistem onu ekrana getirdi 
film.iloc[:,2].head()
```

```python
#0dan 5e kadar olan satirlarin 0dan 3e kadae olan sutunlarinin indexleri yazildi 
film.iloc[0:5,0:3]
```

```python
#0dan 5e kadar olan satirlarin tum suutnlari 
film.iloc[0:5, :]
```

```python
#coklu filtreleme
#hem 200dk dan uzun, hem Crime olan filmler
film [(film.duration>=200)&(film.genre=="Crime")]
```

```python
#200dk dan uzun veya crime 
film [(film.duration>=200)|(film.genre=="Crime")]
```

```python
#birden fazla ayni sutundan secme yapmak icin "isin" methodu kullaniyoruz
film[film.genre.isin(["Action","Crime","Adventure"])]
```
