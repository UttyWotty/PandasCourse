```python
import pandas as pd; import numpy as np
```

```python
veri = pd.DataFrame(np.arange(16).reshape(4,4), index = [list("aabb"),[1,2]*2],
                    columns = [["say","say","soz","soz"],
                               ["mat","ing"]*2])
veri
```

```python
#satir ve sutunlara isim verelim 
#buna genis ya da tablo denir 
veri.index.names= ["sinif","sinav"]
veri.columns.names = ["alan","ders"]
veri
```

```python
#tabloyu uzun formata cevirebiliriz(csv dosyasi gibi ) 
#bunun icin stack , geri almak icin unstack kullanilir
uzun = veri.stack()
uzun
```

```python
 uzun.unstack()
```

```python
#bunu istedigimiz sekilde manipule edebiliriz 
#icine 0 yazdik ve alan degerine gore degistirdi, yada koslei parantez icinde "alan" da yazabiliriz
#stack yaparken eksik veirler gorunmez o yuzden dropna=False kullanilabilie
veri.stack("alan")
```

```python
#tablo formatina cevirme 
#bu database i uzun olarak yaptik 

stok = pd.DataFrame({"meyve":["elma","erik","uzum"]*2,
                     "renk":["mor","sari"]*3,
                     "adet":[3,4,5,6,1,2]})
stok
```

```python
#pivot methoduyla uzun formattaki veriyi tabloya cevirdik
pivot_table = stok.pivot(index="meyve", columns="renk", values="adet")
pivot_table
```

```python
#1 tane daha random deger ekledik 
stok["deger"]=np.random.randn(len(stok))
stok
```

```python
p = stok.pivot(index = "meyve",columns= "renk")
p
```

```python
#deger degiskenine gore alt kumesini bulduk 
p["deger"]
```

```python
#simdi de melt islemi ile tabloyu uzun formata ceviriyoruz 
veri = pd.DataFrame({"ders":["mat","fiz","edb"],
                     "eda":[50,60,70],
                     "efe":[80,70,90],
                     "ali":[60,70,85]})
veri
```

```python
#bu tabloyu uzun formaya ceviriyoruz
grup = pd.melt(veri, ["ders"])
grup
```

```python
data = grup.pivot(index = "ders",columns = "variable",values= "value")
data
```

```python
#dersler suan index satirinda, onu oradan cikaralim 
data.reset_index()
```
